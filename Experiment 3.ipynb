## Problem 1
##### Objective: To show the first five and last five entries of the given dataframe

### Main Sample
##### The approach of this code is to filter the dataframe using the .index attribute for logical opration. By default, .index will produce the range or number of rows of the dataframe therefore it can be exploited such that one can set which rows to print using logical operators. In this case, setting the cars.index < 5 will produce the first 5 rows. We can then append this to a logical or operator so that it can also make an argument for the last 5 rows. In this case, instead of manually loading the dataframe to see which index the program is supposed to start from, we can use the .shape attribute then call index 0 to produce the scalar number of rows. From this, we subtract the desired number of rows starting from the last, which in this case was also 5, then it starts producing the indeces greater and equal to that number.

```python
cars[(cars.index < 5) | (cars.index >= cars.shape[0]-5)]
```

## Problem 2
##### Objective: Extract the following information using subsetting, slicing, and indexing operations.

### a. Display the first five rows with odd-numbered columns (columns 1, 3, 5, 7...) of cars.
##### To achieve the objective, I used .iloc and index splicing to get the first 5 rows. For the columns, i started at index 1 with an increment of 2 to get the odd index of columns.

```python
cars.iloc[0:5, 1::2]
```

### b. Display the row that contains the ‘Model’ of ‘Mazda RX4’
##### To achieve the objective, I used .loc to specify the conditions that must be met to display the row. In this case, the term "Mazda RX4" must be under the column "Model" to satisfy the conditions. This return the columns associated to that model.

```python
cars.loc[cars["Model"] == "Mazda RX4"]
```

### c. How many cylinders (‘cyl’) does the car model ‘Camaro Z28’ have?
##### The same code structure in the previous objective for this one. However, Camaro Z28 was replaced for the term value and a column filter was added such that it only displays the model and the cyl columns.

```python
cars.loc[cars["Model"] == "Camaro Z28", ["Model","cyl"]]
```

### d. Determine how many cylinders (‘cyl’) and what gear type (‘gear’) do the car models ‘Mazda RX4 Wag’, ‘Ford Pantera L’ and ‘Honda Civic’ have.
##### The same code structure in the previous objective for this one. However, there are multiple term values that must be looked up first. By using the logical OR (|) operator, we can iterate the the model name condition for the different brands such that it returns the row value associated with it if it exists in the model name column. Then, it filters the column such that the model, cyl, and gear columns are displayed.


```python
cars.loc[(cars["Model"] == ("Mazda RX4 Wag"))|(cars["Model"] == ("Ford Pantera L"))|(cars["Model"] == ("Honda Civic")), ["Model","cyl","gear"]]
```
